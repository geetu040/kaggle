```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_blobs

# Generate a random dataset
X, y = make_blobs(n_samples=100, n_features=10, centers=6, random_state=42)

print("X.shape", X.shape)
print("y.shape", y.shape)
```

```python
# Standardize the Data
X = (X - X.mean(axis=0)) / X.std(axis=0)
```

```python
# Calculate covariance matrix
cov_mat = np.cov(X, rowvar=False)

# Visualize covariance matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cov_mat, annot=True)
plt.title('Covariance Matrix')
plt.show()
```

```python
#Calculating Eigenvalues and Eigenvectors of the covariance matrix
eigen_values , eigen_vectors = np.linalg.eigh(cov_mat)
```

```python
#sort the eigenvalues in descending order
sorted_index = np.argsort(eigen_values)[::-1]
 
sorted_eigenvalue = eigen_values[sorted_index]
#similarly sort the eigenvectors 
sorted_eigenvectors = eigen_vectors[:,sorted_index]
```

```python
# Draw Scree Plot
plt.bar(range(len(sorted_eigenvalue)), sorted_eigenvalue)
```

```python
# select the first n eigenvectors, n is desired dimension
# of our final reduced data.

# from the scree graph it is obvious that first 3 components are the most valueable
n_components = 3 
eigenvector_subset = sorted_eigenvectors[:,0:n_components]
```

```python
# Transform the data
X_reduced = np.dot(eigenvector_subset.transpose(),X.transpose()).transpose()
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
ax.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2], c=y)
```
